## Writing re-usable code (bonus)

**<a id="#">Task 2.3.4.</a>** Finish the implementation of the function `clip_image` below, by replacing the comment line (the one which starts with `#`) with your code:

```python
#img4=img.copy()

def clip_image(img, T1, T2):
    img_ = img.copy()
    for i in numpy.ndindex(img_.shape[0],img_.shape[1]):
        if img_[i]<T1:
            img_[i]=T1
        elif img_[i]>T2:
            img_[i]=T2
        else: img_[i]=img_[i]
       
    return img_
img4 = clip_image(img,0.31,0.59)
plt.imshow(img4)
plt.colorbar()

```
